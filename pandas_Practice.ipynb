# PANDAS DATASTRUCTURE

Pandas have two types of data structures:
SERIES :- It’s a one dimentional array with indexes, It store single column or row in DataFrame.
DATAFRAME :- It's a tabular spreadsheet which contains one or multiple column or rows.
A one-dimensional array(labelled) capable of holding any one type of data - SERIES
A two-dimentional array(labelled) holding different types of data - DATAFRAMES

## Pandas value_counts()

Return a Series containing counts of unique values.

```python
Age_Dist = df['Age'].value_counts()
Age_Dist
```

###  With normalize set to True, returns the relative frequency by dividing all values by the sum of values.

```python
Age_Dist = df['Age'].value_counts(normalize=True)
Age_Dist
```

## bins

Bins can be useful for going from a continuous variable to a categorical variable; instead of counting unique apparitions of values, divide the index in the specified number of half-open bins.

```python
Age_Dist.value_counts(bins=3)
```

## dropna

With dropna set to False we can also see NaN index values.



```python
Age_Dist.value_counts(dropna=False)
```

```python
Age_Dist.plot(kind = 'bar')
```
