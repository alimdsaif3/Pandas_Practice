```python
## MERGING DATA ######
## Reading dataset about network consumption by ABC company
ABC=pd.read_csv("ABC_company_phone_data.csv")

## Reading TRAI (Telecom Regulatory Authority of India)
TRAI=pd.read_csv("TRAI_data.csv")
```

```python
## Merging Data-2    -Viewing sample data from both datasets and column names
## List function gives you the names of columns
## df.columns function also gives you the columnnames  (df is your dataframe name - like ABC.columns())
print(ABC.head(5))
print(TRAI.head(5))
print(list(ABC), len(ABC))
print(list(TRAI), len(TRAI))
```

```python
print(ABC.columns)
```

```python
## Merge using pandas -  Left outer join, wher ABC is the left table and all 388 records come in merged
Merged_data=pd.merge(ABC, TRAI,on='network', how='left')
print(Merged_data.head(5))
print(len(Merged_data))
```

```python
## Now Describe the merged_data,  by default describe will take only numeric columns, 
## so if you want description of categorical columns, use the swith include='all'

merged_stats_all=Merged_data.describe(include='all')
print(merged_stats_all)

## Date, Network are categorical columns, so no numerical stats shown, only freq and count are shown
## duration and consumption are numeric columns, so all numeric stats are given,  so these are shown as NaNs
```

```python
## Based on above,

## 1) Duration column has a big variance, std dev is high, min is 1, max is 45000, has major outliers
## 2) Consumption is somewhat normally distributed, but count has only 346 values, the rest (388-346) = 42 are missing values

## Outlier treatment for duration column
## lower end outlier is Q1-1.5 IQR = anything below is outlier
## Higher end outlier is Q3+1.5 IQR = anything above this is outlier

## check the column to have a deeper look at outliers - this function will display all the below percentiles
pd.DataFrame(Merged_data['duration']).describe(percentiles=(1, 0.99, 0.9, 0.75, 0.5, 0.3, 0.1,  0.01))
```

```python
## Outlier treatment starts -  InterQuartileRange IQR 

# Convert 'duration' column data to temporary pandas dataframe so that it is easier to apply functions
duration_col_df=pd.DataFrame(Merged_data['duration'])

# Find mean of the 'duration column'  -  int is the integer function - so mean will be returned in integer and not decimals
duration_mean=int(duration_col_df.mean())

# Find 75th percentile of column duration
IQR_duration_P75=duration_col_df.quantile(q=0.75)

#Find 25th percentile of column duration
IQR_duration_P25=duration_col_df.quantile(q=0.25)

# Find IQR for the duration column
duration_IQR=IQR_duration_P75-IQR_duration_P25

##Fix LL and UL boundries to detect outliers in duration column
IQR_LL=int(IQR_duration_P25 - 10*duration_IQR)   # usually its 1.5 times, here since data has huge variance we are using 10times
IQR_UL=int(IQR_duration_P75 + 10*duration_IQR)

print(IQR_LL, IQR_UL, duration_mean)
```

```python
## now we treat the upper end and lower end outliers with 1st percentile and 99th perentile values
Merged_data.loc[Merged_data['duration']<IQR_LL, 'duration']=int(duration_col_df.quantile(q=0.01))
Merged_data.loc[Merged_data['duration']>IQR_UL, 'duration']=int(duration_col_df.quantile(q=0.99))

## loc functions takes rows and cols as arguments,
## the first is a condition for rows where duration is >LL or UL
## the second is the duration column which needs the change
## substite those values with 99th or 1st percentile
```

```python
## Now missing value treatment for Total consumption column
## we will impute mean in place of missing values
Merged_data['Total consumption'].fillna((Merged_data['Total consumption'].mean()), inplace=True)

## Merged data - we take only Total consumption column
## fillna is the function that fills all na values
## fill it with Mean of Merged_data Total consumption column
## inplace =True overwrites existing value
## if you say inplace=False, you have to create a copy of that dataset 
```

```python
## check final stats again after outlier removal and missing value treatment
merged_data_final_stats=Merged_data.describe(include='all')
print(merged_data_final_stats)
```

```python
## Data manipulation - each row consumption as a percentage of total consumption 

# Create a new KPi column called Percentage_of_total for each call
Merged_data['Percentage_of_total']=(Merged_data['duration']*100)/Merged_data['Total consumption']
Merged_data.head(10)
```

```python
## now, we have the percentage for each call duration
## we have to sum up this call duration, for each network provider

Merged_data.groupby(['network'])['Percentage_of_total'].aggregate('sum')

## which network is getting max business from ABC company? -Vodafone, bcoz 7.3% of ABC consumption is from Vodafone network
```
